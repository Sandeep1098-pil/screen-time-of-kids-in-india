**Checking outliers using IQR method**

```python
def count_outliers_iqr(df, column):
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR
    outliers = df[(df[column] < lower) | (df[column] > upper)]
    print(f"{column} â†’ {len(outliers)} outliers")
    return outliers


count_outliers_iqr(df, 'Avg_Daily_Screen_Time_hr')
count_outliers_iqr(df, 'Age')
count_outliers_iqr(df, 'Educational_to_Recreational_Ratio')
```

we can see there are 448 outliers in Avg_Daily_Screen_Time_hr attribute

```python
#drop outliers

def drop_outliers_iqr(df, column):
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR
    return df[(df[column] >= lower) & (df[column] <= upper)]
df = drop_outliers_iqr(df, 'Avg_Daily_Screen_Time_hr')
df.shape
```

Feature Engineering

```python
#converting male and female into 0 and 1
df['Gender_binary'] = df['Gender'].map({'Male': 1, 'Female': 0})
```

```python
#creating attribute urban and rural into binary
df['Urban_binary'] = df['Urban_or_Rural'].map({'Urban': 1, 'Rural': 0})
df
```

```python
#creating attribute with 0 and 1 values by Exceeded_Recommended_Limit
df['Exceeded_Recommended_Limit_binary'] = df['Exceeded_Recommended_Limit'].map({True: 1, False: 0})
df
```
